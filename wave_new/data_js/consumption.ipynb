```python
import sqlite3
import random
from datetime import datetime, timedelta

# Create an array of user IDs
users = [1, 2, 3, 4, 5]

# Generate dummy time-series data for each user
data = []
start_date = datetime.now()  # Use the current date as the starting point

for user_id in users:
    user_series = []
    for i in range(24):
        timestamp = start_date + timedelta(hours=i)  # Add 1 hour to the previous timestamp
        value = random.randint(0, 100)  # Generate a random value

        user_series.append((user_id, timestamp, value))
    data.extend(user_series)
data
```

```python
# Create a SQLite database and insert the data
conn = sqlite3.connect('data.db')
cursor = conn.cursor()

# Create the "data" table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        userId INTEGER,
        timestamp TEXT,
        value INTEGER
    )
''')

# Insert the data into the "data" table
cursor.executemany('''
    INSERT INTO data (userId, timestamp, value)
    VALUES (?, ?, ?)
''', data)

conn.commit()
print('Data inserted into the database.')

# Close the database connection
conn.close()
print('Database connection closed.')
```
